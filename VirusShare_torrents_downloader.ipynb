## VirusShare Torrents Downloader v0.1

```python
%%time

from bs4 import BeautifulSoup, SoupStrainer
import requests
import re
import os, sys
from urllib.parse import urlparse
import urllib.request
from urllib.error import HTTPError
import hashlib
import random
import string
from urlextract import URLExtract
import ssl
import urllib.request

class colors:
    BLUE = '\033[94m'
    GREEN = '\033[92m'
    RED = '\033[91m'
    WARNING = '\033[93m'
    FAIL = '\033[91m'
    ENDC = '\033[0m'

print(colors.GREEN + 'Beginning files downloading...' + colors.ENDC)

def get_data(name):
    data_list = []
    file_name = name
    if not os.path.exists(file_name):
        return 0

    inF = open(file_name, "r") 
    lines = inF.readlines()
    for line in lines:
        data = line.strip()
        if len(data) > 1:
            data_list.append(data)
        else:
            data_list.append(sys.argv[1])
    return data_list

all_ruls = get_data("Z:/temp/virusshare_.txt")

i = 0
for f in range(len(all_ruls)):
    link = all_ruls[f]
    a = urlparse(link)
    filename = os.path.basename(a.path)
    root_dir = 'C:/training/vxshares/' + filename

    print(colors.GREEN +"\n%d. url %s downloading to %s" % (i, link, root_dir)+ colors.ENDC)
    
    try:
        ctx = ssl.create_default_context()
        ctx.check_hostname = False
        ctx.verify_mode = ssl.CERT_NONE

        with urllib.request.urlopen(link, context=ctx) as u, \
                open(root_dir, 'wb') as file:
            file.write(u.read())

            file.close
    except: 
        print (colors.RED + "\n%s download error" % link + colors.ENDC)

    
    i = i + 1

print('DONE!')
```
